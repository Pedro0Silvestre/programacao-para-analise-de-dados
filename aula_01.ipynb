```python
num_funcs = int(input("digite o numero de funcionarios"))

with open('teste.txt', 'w') as f:
    for i in range(1,num_funcs + 1):
        func = input("digite o nome do funcionario:")
        msg = f"{i} - {func}\n"
        f.write(msg)
        
```

```python
import csv

with open('teste.csv', 'w') as f:
    escritor = csv.writer(f)
    escritor.writerow(['Nome','Idade'])
    escritor.writerow(['Pedro', 23])
    escritor.writerow(['miguel', 20])
```

```python
with open('teste.csv', 'r') as f:
    leitor = csv.reader(f)
    for linha in leitor:
        print(linha)
```

```python
try:
    with open('vendas.txt', 'x'):
        pass
except:
    print("arquivo já existe")
```

```python
num_venda = 0
with open('vendas.txt', 'w') as f:
    while True:
        num_venda +=1
        prod = input("insira o produto vendido")
        valor = float(input('insira o valor da venda'))
        f.write(f"venda numero {num_venda}: {prod} - {valor}")
        parada = input("Deseja continuar S/N")
        if parada == 'N':
            break
```
